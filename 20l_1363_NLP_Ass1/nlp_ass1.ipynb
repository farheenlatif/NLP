## Q2) Write regular expressions to match the following classes of strings:

#### a. A single determiner (assume that a, an, and the are the only determiners).
#### b. An arithmetic expression using integers, addition, and multiplication, such as 2*3+8.

```python
import re

# Regular expressions for the specified classes
regexes = {
    "Single_determiner": r"\b(a|an|the)\b",  # Single determiner (a, an, the)
    "Arithmetic_expression": r"\d+(\*\d+|\+\d+)*" # Arithmetic expression with integers, addition, and multiplication
}

# Test strings
test_strings = [
    "The quick brown fox jumps over the lazy dog.",
    "A cat is sleeping on the windowsill.",
    "An orange is a type of fruit.",
    "Please solve the equation: 3x^2 + 2x - 7 = 0.",
    "Can you find the area of a circle with radius 5?",
    "His favorite number is 42.",
]

# Function to test each regular expression
def test_regex(regex, test_strs):
    results = []
    for test_str in test_strs:
        match = re.search(regex, test_str)
        results.append((test_str, bool(match)))
    return results

# Testing each regular expression
test_results = {}
for key, regex in regexes.items():
    test_results[key] = test_regex(regex, test_strings)

# Print results
for key, results in test_results.items():
    print(f"Results for regex '{key}':")
    for test_str, matched in results:
        print(f"  '{test_str}': {'Matched' if matched else 'Did not match'}")
```

## Q3) Beautiful Soup

```python
from urllib import request
from bs4 import BeautifulSoup

def fetch_url_content(url):
    # Fetch the content of the URL
    response = request.urlopen(url)
    html = response.read().decode('utf8')

    # Use BeautifulSoup to remove HTML tags and get plain text
    soup = BeautifulSoup(html, 'html.parser')
    text = soup.get_text()

    return text

# Example Usage
url = 'https://lhr.nu.edu.pk/faculty/'
content = fetch_url_content(url)
print(content)
```

## Q4) Tokenization

```python
def find_phone_numbers(text):
    # Regular expression to match phone numbers
    phone_regex = r'\(\d{3}\) \d{3} \d{3} \d{3}'
    phone_numbers = re.findall(phone_regex, text)
    # Remove any newline characters from phone numbers
    phone_numbers = [number.strip() for number in phone_numbers]
    return phone_numbers

def find_email_addresses(text):
    # Regular expression to match email addresses
    email_regex = r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}'
    email_addresses = re.findall(email_regex, text)
    return email_addresses



# Find phone numbers
phone_numbers = find_phone_numbers(content)
print("Phone Numbers:", phone_numbers)

# Find email addresses
email_addresses = find_email_addresses(content)
print("Email Addresses:", email_addresses)
```

## Q)5 Use the Porter Stemmer and Lancaster Stemmer

```python
# Function to normalize text using Porter Stemmer
def normalize_with_porter_stemmer(tokens):
    porter_stemmer = PorterStemmer()
    stemmed_tokens = [porter_stemmer.stem(token) for token in tokens]
    return stemmed_tokens

# Function to normalize text using Lancaster Stemmer
def normalize_with_lancaster_stemmer(tokens):
    lancaster_stemmer = LancasterStemmer()
    stemmed_tokens = [lancaster_stemmer.stem(token) for token in tokens]
    return stemmed_tokens
```

```python
def tokenize_text_normalize_stemmers(text):
    # Tokenize the text
    tokens = word_tokenize(text)

    # Normalize tokens using Porter Stemmer
    porter_stemmer = PorterStemmer()
    porter_stemmed_tokens = [porter_stemmer.stem(token) for token in tokens]

    # Normalize tokens using Lancaster Stemmer
    lancaster_stemmer = LancasterStemmer()
    lancaster_stemmed_tokens = [lancaster_stemmer.stem(token) for token in tokens]

    return porter_stemmed_tokens, lancaster_stemmed_tokens
```

```python
# Fetch content and tokenize
content = fetch_url_content(url)
tokens = word_tokenize(content)
porter_stemmed_tokens, lancaster_stemmed_tokens = tokenize_text_normalize_stemmers(content)

print("\nOriginal Text: ",tokens)
print("\nPorter Stemmer Tokens:", porter_stemmed_tokens[:100])  # Print first 100 tokens
print("\nLancaster Stemmer Tokens:", lancaster_stemmed_tokens[:100])  # Print first 100 tokens
```
